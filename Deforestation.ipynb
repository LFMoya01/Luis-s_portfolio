<font size="5"> Filter the data for Chile and check for null values

```python
chile_data = carbon[carbon['country'] == 'Chile']
```

<font size="5"> I will only consider a forest canopy threshold of 75 percent to get more representative information about the carbon emissions because aproximetly half of the country is a desert or semi-desert area with little or no considerable forest.

```python
chile_thresholds_data = carbon[(carbon['country'] == 'Chile') & (carbon['umd_tree_cover_density_2000__threshold'].isin([75]))]
```

```python
chile_thresholds_data
```

```python
chile_thresholds_data.isnull().sum()
```

<font size="5"> Based on previous data a new table was created in order to get temporal information for a line chart.

```python
# Extract relevant columns for creating the new DataFrame
emissions_columns = [col for col in chile_thresholds_data.columns if 'gfw_forest_carbon_gross_emissions' in col]

# Create a new DataFrame
new_table = pd.DataFrame()

# Extract year and corresponding carbon gross emissions for each column
for col in emissions_columns:
    year_str = col.split('__')[0][-4:]
    try:
        year = int(year_str)
    except ValueError:
        # Handle columns that do not follow the expected format
        continue
    
    emissions = chile_thresholds_data[col]
    new_table = pd.concat([new_table, pd.DataFrame({'year': [year], 'Carbon gross emissions': [emissions.values[0]]})], ignore_index=True)

# Sort the new DataFrame by 'year'
new_table.sort_values(by='year', inplace=True)

# Display the new DataFrame
print(new_table)
```

<font size="5"> Line chart of the evolution of carbon gross emitions in Chile, every dot equals one year.

```python
# Replace 'year' and 'Carbon gross emissions' with the actual column names

# Plot line chart
plt.figure(figsize=(10, 6))
plt.plot(new_table['year'], new_table['Carbon gross emissions'], marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Carbon Gross Emissions (Mg_CO2e)')
plt.title('Evolution of Carbon Gross Emissions Over Time (Chile)')
plt.grid(True)
plt.show()
```

<font size="5"> The previous process is replicated for the "Tree loss" data in order to get the temporal evolution of the data for Chile. The forest canopy threshold of 75 percent is still the only row considered for the analysis.

```python
chile_treeloss_data = tree_loss[tree_loss['country'] == 'Chile']
```

```python
chile_treeloss_data
```

```python
# Replace 'tc_loss_ha_YYYY' with the actual column name format

# Filter rows where "threshold" is 75
filtered_data = chile_treeloss_data[chile_treeloss_data['threshold'] == 75]

# Extract relevant columns for creating the new DataFrame
loss_columns = [col for col in filtered_data.columns if 'tc_loss_ha' in col]

# Create a new DataFrame
new_loss_table = pd.DataFrame()

# Extract year and corresponding loss ha for each column
for col in loss_columns:
    year_str = col.split('_')[3]
    try:
        year = int(year_str)
    except ValueError:
        # Handle columns that do not follow the expected format
        continue
    
    loss_ha = filtered_data[col]
    new_loss_table = pd.concat([new_loss_table, pd.DataFrame({'year': [year], 'loss ha': [loss_ha.values[0]]})], ignore_index=True)

# Sort the new DataFrame by 'year'
new_loss_table.sort_values(by='year', inplace=True)

# Display the new DataFrame
print(new_loss_table)
```

<font size="5"> After the data processing, the line chart is made.

```python
# Plot line chart
plt.figure(figsize=(10, 6))
plt.plot(new_loss_table['year'], new_loss_table['loss ha'], marker='o', linestyle='-', color='green')
plt.xlabel('Year')
plt.ylabel('Tree Loss (ha)')
plt.title('Evolution of Tree Loss Over Time (Chile - Threshold 75)')
plt.grid(True)
plt.show()
```

<font size="5"> Now both charts, from different dataframes are merged in order to visualize the correlations.

```python
# Create a figure and primary y-axis for Carbon Emissions
fig, ax1 = plt.subplots(figsize=(12, 8))
ax1.plot(new_table['year'], new_table['Carbon gross emissions'], marker='o', linestyle='-', label='Carbon Emissions (Mg_CO2e)', color='blue')
ax1.set_xlabel('Year')
ax1.set_ylabel('Carbon Emissions (Mg_CO2e)', color='blue')
ax1.tick_params('y', colors='blue')
ax1.legend(loc='upper left')

# Create a secondary y-axis for Tree Loss
ax2 = ax1.twinx()
ax2.plot(new_loss_table['year'], new_loss_table['loss ha'], marker='o', linestyle='-', label='Tree Loss', color='green')
ax2.set_ylabel('Tree Loss (ha)', color='green')
ax2.tick_params('y', colors='green')
ax2.legend(loc='upper right')

# Title and grid
plt.title('Carbon Emissions and Tree Loss Over Time (Chile)')
plt.grid(True)

# Show the plot
plt.show()
```

<font size="5"> As is seen on the chart, until year 2020 there is a strong correlation between carbon emissions and tree loss. The peaks are in the same years, 2012 and 2017. Also there is a point between years 2014 and 2015 where, focusing only on the scale, the carbon emissions surpassed the tree losses. Finally, after year 2020 there is an inflextion point where the carbon emissions does not follow the same tendency as tree loss.
