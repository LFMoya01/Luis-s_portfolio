```python
# Assuming 'chile_thresholds_data' is your DataFrame
# Replace 'gfw_forest_carbon_gross_emissions_YYYY__Mg_CO2e' with the actual column name format

# Extract relevant columns for plotting
emissions_columns = [col for col in chile_thresholds_data.columns if 'gfw_forest_carbon_gross_emissions' in col]
plot_data = chile_thresholds_data[emissions_columns]

# Plot line chart
plt.figure(figsize=(10, 6))

for col in emissions_columns:
    plt.plot(plot_data.index, plot_data[col], label=col)

plt.xlabel('Index')
plt.ylabel('Carbon Gross Emissions (Mg_CO2e)')
plt.title('Carbon Gross Emissions Over Time (Chile)')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
# Replace 'gfw_forest_carbon_gross_emissions_YYYY__Mg_CO2e' with the actual column name format

# Extract relevant columns for creating the new DataFrame
emissions_columns = [col for col in chile_thresholds_data.columns if 'gfw_forest_carbon_gross_emissions' in col]

# Create a new DataFrame
new_table = pd.DataFrame()

# Extract year and corresponding carbon gross emissions for each column
for col in emissions_columns:
    year_str = col.split('__')[0][-4:]
    try:
        year = int(year_str)
    except ValueError:
        # Handle columns that do not follow the expected format
        continue
    
    emissions = chile_thresholds_data[col]
    new_table = pd.concat([new_table, pd.DataFrame({'year': [year], 'Carbon gross emissions': [emissions.values[0]]})], ignore_index=True)

# Sort the new DataFrame by 'year'
new_table.sort_values(by='year', inplace=True)

# Display the new DataFrame
print(new_table)
```

```python
new_table
```

```python
# Replace 'year' and 'Carbon gross emissions' with the actual column names

# Plot line chart
plt.figure(figsize=(10, 6))
plt.plot(new_table['year'], new_table['Carbon gross emissions'], marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Carbon Gross Emissions (Mg_CO2e)')
plt.title('Evolution of Carbon Gross Emissions Over Time (Chile)')
plt.grid(True)
plt.show()
```
